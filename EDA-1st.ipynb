```python
df_country =pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df =pd.merge(df ,df_country , on='Country Code' ,how='left')
```

```python
final_df.head(2)
```

```python
# now to check the datatypes
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# pie chart --top 3 countries that uses zomato or the online transactions
plt.pie(country_val[:3], labels=country_names[:3])
```

Now we are clear that maximum number of online transactions are happening in India itself and then in united states and then in united kingdom

```python
# pie chart --top 3 countries that uses zomato or the online transactions
plt.pie(country_val[:3], labels=country_names[:3] ,autopct='%1.2f%%')
```

Observation: Zomato maximum records and transaction are fromIndia After that United States and then United Kingdom

renaming the 0 column to Rating count

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating =final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

**##Observations**
1. when rating is between 4.5 to 4.9 ---->excellent rating 
2. when rating is between 4.0 to 4.4 ----->Very Good
3. when rating is between 3.4 to 3.9 ----->Good
4.  when rating is between 2.5 to 2.9 ----> Average
5. when rating is between 1.8 to 2.4 ----> poor

```python
rating.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6 )
sns.barplot(x="Aggregate rating", y="Rating count" , data=rating)
```

we should also show the colors as per them mentioned in the rating.

```python
sns.barplot(x="Aggregate rating", y="Rating count" , hue='Rating color' , data=rating)
```

now mapping the colors


```python
sns.barplot(x="Aggregate rating", y="Rating count" , hue='Rating color' , data=rating ,palette=['white', 'red','orange','yellow','green','green'])
```

Observaton :
1. Not rated count is very high
2. Maximum no of rating is between 2.5 to 3.4

```python
# count plot
sns.countplot(x='Rating color',data =rating, palette=['blue', 'red','orange','yellow','green','green'])
```
